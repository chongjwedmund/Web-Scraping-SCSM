```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import requests
import re
```

```python
url = 'https://www.sportsplits.com/races/singapore-marathon-2022/events/1/results/individuals/'
```

```python
# function for web scraping and initial data cleaning
def get_page(page):
    url
    url_next = url+f'{page}'
   
    resp = requests.get(url_next, verify=False)   
    soup = BeautifulSoup(resp.text) # make soup 
    tables = soup.find_all('table') 
    tr_list = tables[0].find_all('tr')
    data = []

    for tr in tr_list:
        td_list = tr.find_all('td') # find all table data and append each row as an element in a list 
        row = [td.text for td in td_list]
        if row:
            data.append(row)
    
    stripper = [1] # strip remaining cols of useless characters 
    for i in range(len(data)):
        for k in stripper: 
            data[i][k] = data[i][k].replace('\n','').replace('\t','').replace('|','')
            
    # tag race number with rows 
    for i in range(len(data)): 
        race_num = page
        if race_num:
            data[i].append(page)

    return(data)
```
