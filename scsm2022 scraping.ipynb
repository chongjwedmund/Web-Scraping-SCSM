```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import requests
import re
```

```python
url = 'https://www.sportsplits.com/races/singapore-marathon-2022/events/1'
```

```python
# function for web scraping and initial data cleaning
def get_page(page):
    url
    url_next = url+f'?page={page}'
   
    resp = requests.get(url_next, verify=False)   
    soup = BeautifulSoup(resp.text) # make soup 
    tables = soup.find_all('table') 
    tr_list = tables[0].find_all('tr')
    data = []

    for tr in tr_list:
        td_list = tr.find_all('td') # find all table data and append each row as an element in a list 
        row = [td.text for td in td_list]
        if row:
            data.append(row)
            
    dels = [3,8] # delete unused columns
    for i in range(len(data)):
        for j in sorted(dels, reverse = True): 
            del data[i][j] 
    
    stripper = [3,4,5] # strip remaining cols of useless characters 
    for i in range(len(data)):
        for k in stripper: 
            data[i][k] = data[i][k].replace('\n','').replace('\t','').replace('|','').replace('\xa0','')

    return(data)
```

```python
# test the function 
get_page(1)[:5]
```

```python
pages = list(range(155)) # there are 155 pages
result = []

for page in pages: # combining all pages 
    result.append(get_page(page))
    
flat_result = [item for sublist in result for item in sublist] # flattening nested lists to return single list 
flat_result[:5]
```

```python
header = ['Pos', 'Name', 'Gun Time', 'Country', 'Category (Pos)', 'Gender (Pos)', 'Net Pos']
```

```python
with open('scsm2022.csv','w', newline='', encoding="utf-8") as f: 
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(flat_result)
```
